```python
from dataloader import DataLoader
dataloader = DataLoader(base_path="./datasets")
```

```python
DATASET_NAME = "rest15"
```

```python
dataset = [] 
for ds_name in ["train", "test", "dev"]:
  dataset += dataloader.load_data(name=DATASET_NAME, data_type=ds_name)
len(dataset)
```

```python
import random 
import numpy as np
from sklearn.preprocessing import MultiLabelBinarizer
from iterstrat.ml_stratifiers import MultilabelStratifiedKFold

def random_cross_validation_split(data, seed=42):
    categories = [[el["aspect_category"] for el in an["aspects"]] for an in data]

    mlb = MultiLabelBinarizer()
    Y = mlb.fit_transform(categories)
    print(Y)
    category_names = mlb.classes_

    n_splits = 5
    mskf = MultilabelStratifiedKFold(n_splits=n_splits, shuffle=True, random_state=seed)
    
    splits = []
    for train_index, test_index in mskf.split(np.zeros(len(Y)), Y):
       splits.append((train_index, test_index))
    
    
    return splits
```

```python
splits = random_cross_validation_split(dataset)
categories = [[el["aspect_category"] for el in an["aspects"]] for an in dataset]
```

```python
mlb = MultiLabelBinarizer()
Y = mlb.fit_transform(categories)
category_names = mlb.classes_
```

```python
for i, (train_idx, test_idx) in enumerate(splits):
    print(f"Split {i + 1}:\n")

    test_counts = np.sum(Y[test_idx], axis=0)
    
    print("\nTest Data Category Counts:")
    for category, count in zip(category_names, test_counts):
        print(f"{category}: {count}")
    
    print("\n" + "-"*40 + "\n")
```
