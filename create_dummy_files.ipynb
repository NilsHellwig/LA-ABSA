```python
import os
import random
import shutil
```

```python
folder_path = "./_out_synthetic_examples/03_llm_eda_few_shot_augmenter/"

for task in ["asqp", "tasd"]:
    for dataset in ["rest15", "rest16", "hotels", "coursera", "flightabsa"]:
        for n_shot in [10, 50]:
            copy_file = folder_path + f"{task}_rest15_10.txt"
            duplicate_path = os.path.join(folder_path, f"{task}_{dataset}_{n_shot}.txt")
            try:
                shutil.copy2(copy_file, duplicate_path)
            except:
                pass
```

```python
####### Fine tunings

folder_path = "./_out_fine_tunings/01_llm_annotate_train/"

for task in ["asqp", "tasd"]:
    for ml_method in ["paraphrase", "dlo"]:
        for dataset in ["rest15", "rest16", "hotels", "coursera", "flightabsa"]:
            for n_llm_examples in ["full", 800]:
                for fs_num in [0, 10, 50]:
                    for seed in range(5):
                        copy_file = (
                            folder_path + f"paraphrase_full_{task}_50_rest16_0.json"
                        )
                        duplicate_path = os.path.join(
                            folder_path,
                            f"{ml_method}_{n_llm_examples}_{task}_{fs_num}_{dataset}_{seed}.json",
                        )
                        try:
                            shutil.copy2(copy_file, duplicate_path)
                        except:
                            pass

```
